```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("soc-sign-bitcoinotc.csv")
print(df.head())
```

```python
g = nx.DiGraph()
for i,row in df.iterrows():
    a,b,c,d = row
    a = int(a)
    b = int(b)
    if a not in g:
        g.add_node(a)
    if b not in g:
        g.add_node(b)
    g.add_edge(a,b)
    c = int(c)
    g[a][b]["weight"] = c
    g[a][b]["timestamp"] = d
print(g)
```

```python
centralities = nx.closeness_centrality(g)
best = [(value, key) for key, value in centralities.items()]
print ("Closensess centrality\nNode: ",max(best)[1])
print("Value: ",centralities[max(centralities, key=centralities.get)])
```

```python
betweenness = nx.betweenness_centrality(g)
best = [(value, key) for key, value in betweenness.items()]
print ("Betweenness centrality\nNode: ",max(best)[1])
print("Value: ",betweenness[max(betweenness, key=betweenness.get)])
```

```python
#The graph is not connected
print("Max shortest path",max([max(j.values()) for (i,j) in nx.shortest_path_length(g)]))
```

```python
print("Average clustering", nx.average_clustering(g))
```

```python
print("Number of connected components: ",nx.number_weakly_connected_components(g))
```

```python
#There aren't single nodes with zero degree
count=0
for i in g.nodes:
    if (g.in_degree[i]==0 and g.out_degree[i]==0):
        count=count+1
print(count)
```

```python
print(len(list(nx.weakly_connected_components(g))))
biggest_component_nodes = max(nx.weakly_connected_components(g), key=len)
print(len(biggest_component_nodes))
```

```python
#Subgraph connected
sg=nx.subgraph(g, biggest_component_nodes)
print(sg)
```

```python
print("Number of connected components: ",nx.number_weakly_connected_components(sg))
```

```python
centralities = nx.closeness_centrality(sg)
best = [(value, key) for key, value in centralities.items()]
print ("Closensess centrality\nNode: ",max(best)[1])
print("Value: ",centralities[max(centralities, key=centralities.get)])
```

```python
betweenness = nx.betweenness_centrality(sg)
best = [(value, key) for key, value in betweenness.items()]
print ("Betweenness centrality\nNode: ",max(best)[1])
print("Value: ",betweenness[max(betweenness, key=betweenness.get)])
```
